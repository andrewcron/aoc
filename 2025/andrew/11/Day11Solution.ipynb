```python
test = False
filename = "test.txt" if test else "input.txt"
with open(filename) as f:
    lines = f.read().splitlines()
network = {}
for line in lines:
    start, ends = line.split(": ")
    network[start] = ends.split(" ")
```

```python
rev_network = {}
for start, ends in network.items():
    for end in ends:
        rev_network.setdefault(end, set()).add(start)
```

```python
def run_dfs(start="you", end="out", exclude=set()):
    paths = []
    def dfs(node, path):
        if node == end or node == "out":
            paths.append(path + [node])
            return
        for neighbor in network[node]:
            if neighbor in path or neighbor in exclude:
                continue
            dfs(neighbor, path + [node])

    dfs(start, [])
    return set([tuple(p) for p in paths if p[-1] == end])   
```

```python
def find_number_of_paths(start="you", end="out", exclude=set()):
    paths_from = {}
    def dfs(node):
        if node in paths_from:
            return paths_from[node]
        if node == end or node == "out":
            return 1
        total_paths = 0
        for neighbor in network[node]:
            if neighbor in exclude:
                continue
            total_paths += dfs(neighbor)
        paths_from[node] = total_paths
        return total_paths
    return dfs(start)
```

```python
route1 = find_number_of_paths("svr", "fft", {"dac", "out"}) * \
    find_number_of_paths("fft", "dac", {"svr", "out"}) *\
    find_number_of_paths("dac", "out", {"svr", "fft"})
route2 = find_number_of_paths("svr", "dac", {"fft", "out"}) * \
    find_number_of_paths("dac", "fft", {"svr", "out"}) *\
    find_number_of_paths("fft", "out", {"svr", "dac"})
route1 + route2
```
