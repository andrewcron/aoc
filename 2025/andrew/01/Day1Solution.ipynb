```python
mode = "input"  # Change to "input" for production mode

# Set filename based on mode
if mode == "test":
    filename = "test.txt"
else:
    filename = "input.txt"

# Read the file and clean each row into a list
with open(filename, 'r') as f:
    data = [line.strip() for line in f.readlines()]

```

```python
# setup value counter
value_counter = {i: 0 for i in range(0, 100)}

# start at value 50
current_value = 50

# Process each line in the data
for line in data:
    # get the direction and amount
    direction = line[0]
    amount = int(line[1:])
    # update the current value
    if direction == 'L':
        current_value -= amount
    elif direction == 'R':
        current_value += amount
    else:
        raise ValueError(f"Unknown direction: {direction}")
    # increment the counter for this value
    value_counter[current_value % 100] += 1   
    
```

```python
# report how many times we visited value 0
print(f"Value 0 was visited {value_counter[0]} times.")
```

**Part 2, count all the ticks, not just the end of the turn**

```python
# initialize zero visits and current value
zero_visits = 0
current_value = 50
prev_value = current_value

# Process each line in the data
for line in data:
    # get the direction and amount
    direction = line[0]
    amount = int(line[1:])
    multiplier = 1 if direction == 'R' else -1
    # if the amount is large enough to cross zero multiple times
    if amount >= 100:
        zero_visits += amount // 100
        amount = amount % 100

    if amount > 0:
        # update the current value
        current_value += multiplier * amount

        # check if we crossed zero and update zero visits
        if current_value % 100 == 0:
            zero_visits += amount // 100 + 1
        elif prev_value < 100 and current_value > 100:
            zero_visits += amount // 100 + 1
        elif prev_value > 0 and current_value < 0:
            zero_visits += amount // 100 + 1

    # wrap the current value within 0-99
    current_value = current_value % 100
    prev_value = current_value


```

```python
# report how many times we crossed value 0
print(f"Value 0 was crossed {zero_visits} times.")
```
