```python
test = False  # Set to True for test.txt, False for input.txt

filename = "test.txt" if test else "input.txt"
with open(filename, 'r') as f:
    lines = f.readlines()
```

```python
nums = []
nn = 3 if test else 4
for i in range(nn):
    nums.append([int(x) for x in lines[i].strip().split(' ') if x != ''])   
operations = [x for x in lines[nn].strip().split(' ') if x != '']    
   
```

```python
total = 0
for i in range(len(operations)):
    op = operations[i]
    if op == '+':
        subtotal = 0
        for j in range(len(nums)):
            subtotal += nums[j][i] 
    elif op == '*':
        subtotal = 1
        for j in range(len(nums)):
            subtotal *= nums[j][i]
    total += subtotal
print(f"Total sum: {total}")
```

```python
## part 2
# need to process lines differently
nums, new_nums = [], []
operations = []
nn = 3 if test else 4
for i in range(len(lines[-1])):

    if ''.join([lines[j][i] for j in range(nn)]).strip() == '':
        nums.append(new_nums)
        new_nums = []

    num_str = ''.join([lines[j][i] for j in range(nn)])
    if num_str.strip() != '':
        new_nums.append(int(num_str.strip()))

    if lines[-1][i] != ' ':
        operations.append(lines[-1][i])

nums.append(new_nums)
        
```

```python
total = 0
for i in range(len(operations)):
    op = operations[i]
    # print(f"Processing operation {op} on numbers {nums[i]}")
    if op == '+':
        # print("Addition")
        subtotal = 0
        for j in range(len(nums[i])):
            # print(f" Adding {nums[i][j]} to subtotal {subtotal}")
            subtotal += nums[i][j] 
    elif op == '*':
        # print("Multiplication")
        subtotal = 1
        for j in range(len(nums[i])):
            # print(f" Multiplying {nums[i][j]} with subtotal {subtotal}")    
            subtotal *= nums[i][j]
    else:
        subtotal = 0
    total += subtotal
print(f"Total sum: {total}")
```
