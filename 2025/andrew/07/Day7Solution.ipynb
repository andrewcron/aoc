```python
from functools import cache
```

```python
test = False
filename = "test.txt" if test else "input.txt"
with open(filename) as f:
    lines = f.read().splitlines()

grid = [list(line) for line in lines]
```

```python
def print_row(splitters, beams, cols):
    row = ['.']*cols
    for j in beams:
        row[j] = '|'
    for j in splitters:
        row[j] = '^'
    print(''.join(row))
```

```python
rows, cols = len(grid), len(grid[0])
start_col = grid[0].index('S')
sparse_grid = {i: tuple(j for j, val in enumerate(row) if val == '^') for i, row in enumerate(grid)}
```

```python
beams = set((start_col,))
num_splits = 0
for r in range(1,rows):
    # progress beams downwards
    if r in sparse_grid:
        splitters = sparse_grid[r]
        new_beams = set()
        for beam in beams:
            if beam in splitters:
                new_beams.add(beam-1)
                new_beams.add(beam+1)
                num_splits += 1
            else:
                new_beams.add(beam)
        beams = new_beams
        print_row(splitters, beams, cols)
print("----")
print(f"num_splits: {num_splits}")

```

```python
@cache # memoization to speed up recursive calls ... very slow otherwise
def quantum_row_paths(r, c):
    if r == rows - 1:
        return 1
    possibilities = 0
    if r+1 in sparse_grid:
        splitters = sparse_grid[r+1]
        if c in splitters:
            possibilities += quantum_row_paths(r+1, c-1)
            possibilities += quantum_row_paths(r+1, c+1)
        else:
            possibilities += quantum_row_paths(r+1, c)
    else:
        possibilities += quantum_row_paths(r+1, c)
    return possibilities

quantum_row_paths(0, start_col)
```
