```python
# Read input based on test flag
test = False  # Set to True to read from test.txt
filename = "test.txt" if test else "input.txt"

pairs = []
numbers = []

with open(filename, 'r') as f:
    # Read pairs until blank line
    for line in f:
        line = line.strip()
        if not line:
            break
        if '-' in line:
            pair = tuple(map(int, line.split('-')))
            pairs.append(pair)
    
    # Read single numbers after blank line
    for line in f:
        line = line.strip()
        if line:
            numbers.append(int(line))

print(f"Pairs: {pairs}")
print(f"Single numbers: {numbers}")
```

```python
# sort the pairs by their starting number
pairs.sort(key=lambda x: x[0])
# sort the numbers
numbers.sort()
```

```python
# Create a working copy of pairs
working_pairs = pairs.copy()

total = 0
# iterate through each number and check if it falls within any pair
for number in numbers:
    found = False

    # remove pairs that end before the current number
    working_pairs = [pair for pair in working_pairs if pair[1] >= number]
    for start, end in working_pairs:        
        if start <= number <= end:
            found = True
            break
    if found:
        total += 1

print("---")
print(f"Total numbers within pair ranges: {total}")
```

```python
# part 2 - now the pairs are intervals, we need to merge overlapping intervals

merged_pairs = []
unmerged_pairs = pairs.copy()

while unmerged_pairs:
    # take the first unmerged pair
    # print(f"Unmerged pairs: {unmerged_pairs}")
    # print(f"Merged pairs so far: {merged_pairs}")
    current_start, current_end = unmerged_pairs.pop(0)
    merged = False

    # find if it overlaps with any merged pair
    for i, (start, end) in enumerate(merged_pairs):
        # check for any overlap, including touching intervals and partial overlaps
        if not (current_end < start or current_start > end):
            # Overlapping intervals, merge them
            new_start = min(current_start, start)
            new_end = max(current_end, end)
            merged_pairs[i] = (new_start, new_end)
            merged = True
            break


    if not merged:
        merged_pairs.append((current_start, current_end))



```

```python
print(f"Original pairs: {len(pairs)}")
print(f"Total merged pairs: {len(merged_pairs)}")
# print the total number of integers covered by the merged pairs inclusive of endpoints
total_covered = 0
for start, end in merged_pairs:
    total_covered += (end - start + 1)
print(f"Total covered integers: {total_covered}")
```
