```python
test = False  # Set to False to use input.txt

filename = "test.txt" if test else "input.txt"

with open(filename, 'r') as f:
    lines = f.readlines()

data = [[int(char) for char in line.strip()] for line in lines]
```

```python
total = 0
for row in data:
    print(row)
    # get the large=st number in the row and the index excluding the last element
    largest = max(row[:-1])
    index = row.index(largest)
    print(f"Largest number: {largest} at index {index}")
    # get the next largest number in the row after the index of the largest number
    next_largest = max(row[index + 1:])
    print(f"Next largest number after index {index}: {next_largest}")
    # combine the two digits to form a two-digit number
    two_digit_number = largest * 10 + next_largest
    print(f"Two-digit number formed: ***{two_digit_number}***")
    total += two_digit_number
print(f"Total: {total}")
```
